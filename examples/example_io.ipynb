```python
import napari
from napari.utils import nbscreenshot

viewer = napari.Viewer()

# set the viewer in 3D mode
viewer.dims.ndisplay = 3
```

```python
from napari_swc_editor._sample_data import make_sample_data

point_layer, shape_layer = make_sample_data()

nbscreenshot(viewer)
```

```python
# select some points

point_layer.selected_data = list(range(200))

# you can also do it manually on the viewer

nbscreenshot(viewer)
```

```python
# remove points

point_layer.remove_selected()
nbscreenshot(viewer)
```

```python
# save swc

napari.save_layers('test.swc', [point_layer])

# remove all layers
viewer.layers.select_all()
viewer.layers.remove_selected()
```

```python
# load new swc

viewer.open('test.swc', plugin='napari-swc-editor') # currently this will work but throw an error
```
